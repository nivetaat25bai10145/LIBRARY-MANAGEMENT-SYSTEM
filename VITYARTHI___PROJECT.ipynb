```python
import pandas as pd
from datetime import datetime
def get_valid_date(prompt):
    while True:
        date_str = input(prompt)
        try:
            dt = datetime.strptime(date_str, "%Y-%m-%d")
            return dt.strftime("%Y-%m-%d")
        except ValueError:
            print("Invalid date.")

print("\n=====WELCOME TO VIT-BHOPAL- LIBRARY MANAGEMENT====\n")
global df
df=pd.DataFrame(columns=['Name','User_ID'])
def create_account():
    print("* NEW USER CREATE AN ACCOUNT *")
    Name=input("Enter the name of the user:")
    User_ID=input("Enter the user ID:")
    global df
    df.loc[len(df)]=[Name,User_ID]
    print(df.to_string(index=False))
    print("\n| ACCOUNT CREATED SUCCESSFULLY ||\n")
create_account()

df=pd.DataFrame(columns=['Book_id','book_name','Author_name'])
data={ 'book_id': [101, 102, 103, 104, 105, 106, 107, 108],'book_name': ['CALCULAS','COMP_CHEMISTRY', 'EEE', 'EFFECTIVE COMMUNICATION', 'THE_DOWNTOWN_CITY',
        'THE GREAT BRITAN','THE LION KING','THE TALE OF GALLET' ],'author_name': ['B.S. GREWEL','ALEXANDAR','BATTACHARYA','G.V.PRAKASH','GERONIMO',
        'RAMCHARAN','HIT MAN','VK DAS']}
df=pd.DataFrame(data)
df['Status']='Available'
print(df.to_string(index=False))
last_book_id=108

def book():
    global last_book_id
    print("\n=== Add New Book ===")
    book_name=input("Enter The Book Name: ")
    author_name=input("Enter The Author Name: ")
    last_book_id+=1
    add_book(last_book_id, book_name, author_name)
    print("Successfully added")
    print(df.to_string(index=False))

def add_book(book_id, book_name, author_name):
    global df
    new_row=[book_id, book_name, author_name, 'Available']
    df.loc[len(df)]=new_row

def show_all_book():
    print("\n=== Displaying Available Books ===")
    available_books=df[df['Status']=='Available']
    if not available_books.empty:
        print("The Books Present In Library:")
        print(available_books.to_string(index=False))
    else:
        print("No books are currently available.")



def remove_book():
    global df
    ("\n=== Add  ===")
    value_to_delete=input("Enter the Book Name to delete: ")
    df=df[df['book_name']!=value_to_delete].reset_index(drop=True)
    print("Updated DataFrame:")
    print(df.to_string(index=False))



def borrow_a_book():
    global df
    print("\n=== Borrow A New Book ===")
    book_name=input("Enter The Book Name: ")
    book_mask=(df['book_name'] == book_name)

    if book_mask.any():
        if df.loc[book_mask, 'Status'].iloc[0]!='Available':
            print("\n",book_name, "is currently not available for borrowing.")
            return

        print("\n Book Found! Please provide details.")

        borrow_time = get_valid_date("Enter The Book Borrowing Date (YYYY-MM-DD): ")
        due_time = get_valid_date("Enter The Book Due Date (YYYY-MM-DD): ")

        df.loc[book_mask, 'Borrow_Time']=borrow_time
        df.loc[book_mask, 'Due_Time']=due_time
        df.loc[book_mask, 'Status']='Borrowed'

        print("\n YOU BORROWED A BOOK , THANK YOU")

    else:
        print("\nThe Entered Book Is Not Available.")

    print("\n--- UPDATED LIBRARY BOOKS ---")
    print(df[book_mask].to_string(index=False))


def return_book():
    global df
    print("\n===== To Return The Borrowed Book =====\n")
    book_name=input("Enter The Book Name to return: ")
    book_mask=(df['book_name'] == book_name)

    if book_mask.any():
        if df.loc[book_mask, 'Status'].iloc[0]=='Borrowed':
            df.loc[book_mask, 'Status']='Available'
            print("\nBook",book_name,"has been returned successfully!")
        else:
            print("\nBook",book_name, "is not currently borrowed.")
    else:
        print("\nThis book does not exist in the library.")

    print("\n--- UPDATED LIBRARY BOOKS ---")
    print(df.to_string(index=False))


def main_menu():
    while True:
        print("\n=====WELCOME TO VIT-BHOPAL- LIBRARY MANAGEMENT====\n")
        print("1) ADD BOOK")
        print("2) SHOW ALL BOOK")
        print("3) REMOVE BOOK")
        print("4) BORROW A BOOK")
        print("5) RETURN A BOOK")
        print("6) Exit")
        choice =int(input("Enter Your Choice:"))
        if choice ==1:
            book()
        elif choice ==2:
            show_all_book()
        elif choice ==3:
            remove_book()
        elif choice ==4:
            borrow_a_book()
        elif choice ==5:
            return_book()
        elif choice ==6:
            print("\nThank you for using the Library Management System!")
            break
        else:
            print("\nInvalid option. Please select a number from the menu.")

main_menu()
```
